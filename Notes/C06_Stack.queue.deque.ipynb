Stack, queue
Stack and queue are not the data structure that are originally exist in python. We need to implement them by list in python. 
define classes of stack and queue.

stack: last in, first out. 后进先出
S.add(e) implemented by S.append(), add element e at the back of the list;
S.pop() remove the last element

queue: first in, first out. 先进先出
implemented by using an array circularly..
use an explicit variable f to store the index of the element at the front of the queue
update f by f = (f+1)% N when pop the front element out
add new element 

deque: double ended queue; both ends can add and remove element
python has a built-in moudle: collections.deque, can use deque class directly.

```python
class ArrayStack:

    def __init__(self):
        self._data = []
        self._size = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        if self._size == 0:
            return True
        else:
            return False
    
    def add(self,e):
        self._data.append(e)
        self._size += 1
    
    def pop(self):
        if self.is_empty():
            raise Empty("Stack is empty.")
        self._data.pop()
        self._size -= 1
    
    def top(self):
        if self.is_empty():
            raise Empty("Stack is empty.")
        return self._data[-1]
        
```

```python
class ArrayQueue:
    
    DEFAULT_CAPACITY = 10 
    
    def __init__(self):
        """create an empty queue."""
        self._data = [None]*ArrayQueue.DEFAULT_CAPACITY
        self._size = 0
        self._front = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def first(self):
        if self.is_empty():
            raise Empty('queue is empty.')
        return self._front

    def dequeue(self):
        
        """ remove and return the front element of the queue. """
        # dequeue这里就需要改变self._front的值了, self._size也需要改
        
        if self.is_empty():
            raise Empty('queue is empty.')

        answer = self._data[self._front]   # the element of the front
        self._data[self._front] = None     # set the front as None, it's like pop the front
        self._front = (self._front + 1) % len(self._data)    # reset the front 
        self._size -= 1
        return answer
    
    def enqueue(self,e):
        """ add an element to the back of the queue. """
        # 加element，这里需要改self._size值
        # 如果数组已经满了，double the size
        if self._size == len(self._data):
            self._resize(2*len(self._data))
        avail = (self._front + self._size) % len(self._data)
        self._data[avail] = e
        self._size += 1
    
    def _resize(self, cap):
        """ resize the list: create a new list, copy the queue to the new list which start with the front element
        of the old queue. Here, need to set self._front to zero after resizing."""
        old = self._data
        self._data = [None]*cap
        walk = self._front
        for k in range(self._size):
            self._data[k]= old[walk]
            walk = (1+walk)%len(old)
        self._front = 0
        
        
```

```python
class ArrayDeque:
    
    DEFAULT_CAPACITY = 10
    
    def __init__(self):
        self._data=[None]*ArrayDeque.DEFAULT_CAPACITY
        self._size = 0
        self._front = 0
        self._back = self._front + self._size - 1
    
    def __len__(self):
        return self._size
    
    def first(self):
        assert self._size > 0
        return self._data[self._front]
    
    def last(self):
        assert self._size > 0 
        return self._data[self._back]
    
    def reserve(self, cap):
        old = self._data
        self._data = [None] * cap
        walk = self._front
        for k in range(self._size):
            self._data[k] = old[walk]
            walk = (walk + 1) % len(old)
        self._front = 0
        self._back = self._front + self._size - 1
    
    def delete_first(self):
        answer = self._data[self._front]
        self._data[self._front] = None
        self._size -= 1
        self._front = (self._front + 1) % len(self._data)
        return answer
        
    def delete_last(self):
        answer = self._data[self._back]
        self._data[self._back] = None
        self._size -= 1
        self._back = (self._front + self._size - 1) % len(self._data)
        return answer
    
    def add_last(self, e):
        if self._size == len(self._data):
            self.reserve(2 * len(self._data))
        self._back = (self._back + 1) % len(self._data)
        self._data[self._back] = e
        self._size += 1
    
    def add_first(self,e):
        if self._size == len(self._data):
            self.reserve(2*len(self._data))
        self._front = self._back + len(self._data) - self._size
        self._data[self._front] = e 
        self._size += 1
        
    
```
