```python
"""
Write a Python program that outputs 
all possible strings formed by using the characters c , a , t , d , o ,and g exactly once.
"""
from random import choice
def possiblestr():
    l = ['c','a','t','d','o','g']
    n = len(l)
    char = ''
    while len(char) < n:
        lett = choice(l)
        if lett not in char:
            char += lett
    return char

def allstr():
    allstr = []
    i = 0
    while i < 10000:
        strs = possiblestr()
        if strs not in allstr:
            allstr.append(strs)
        i += 1
    return len(allstr), allstr

print(allstr())

    
        
```

```python
"""
Write a Python program that can take a positive integer greater than 2 as input and write out 
the number of times one must repeatedly divide this number by 2 before getting a value less than 2.
"""

def count2divid(n):
    count = 0
    if n < 2:
        return count
    else:
        while n >= 2:
            n = n / 2
            count += 1
        return count

count2divid(50)
            
```

```python
"""
A common punishment for school children is to write out a sentence multiple times. 
Write a Python standalone program that will write out the following sentence one hundred times: 
“I will never spam my friends again.” Your program should number each of the sentences and 
it should make eight different random-looking typos.
"""
import random
def copy(s,n,m):
    """
    input: s --> string
    input: n --> int, copy n times
    input: m --> int, make m typos
    output: string copied n times, with m random-looking typos
    
    ord('a') = 97, ord('z') = 122
    ord(letter)-97 -- 'a' start from 0, then 'z' corresponds to 25
    ord(letter)-97 + b  --- randomly increas b
    (ord(letter)-97 + b)% 25 -- limit the number from 0 to 25
    (ord(letter)-97 + b)% 25 + 97 -- back to ASCII number of letters
    chr((ord(letter)-97 + b)% 25 + 97)  -- change number to letter
    """
    i = 0
    count = 0
    while i < n:
        s_newlist = [0]*len(s)
        a = random.randint(2, len(s)-1)  # the changed position in s except the first two and the last.
        for j in range(0,len(s)):
            if count < m: 
                if j == a and s[j] != ' ':
                    b = random.randint(1, 25)   # the letter at position a add randomly b to change to other random letter
                    s_newlist[j] = chr(((ord(s[j])-97+b) % 25)+97)
                    count += 1
                else:
                    s_newlist[j] = s[j]
            else:
                s_newlist[j] = s[j]
            
        s_news = ''
        for letter in s_newlist:
            s_news += letter    
        print(s_news)       
        i += 1
    print('number of random typos:', count)
    
s = 'I will never spam my friends again.'
n = 20
m = 8
copy(s,n,m)

# 这种方法，每句最多一个typo，一句不会出现多个typos
```

```python
"""
The birthday paradox says that the probability that two people in a room will have the same birthday 
is more than half, provided n, the number of people in the room, is more than 23. This property is not really 
a paradox, but many people find it surprising. Design a Python program that can test this paradox by a series 
of experiments on randomly generated birthdays, which test this paradox for n = 5,10,15,20,...,100.
"""
def testparadox(m,n):
    """
    input: m - int, the number of experiments
    input: n - the number of people 
    output: (0,1) - the probability that same birthday appears in the classroom
    """
    exp = 0
    count = 0      
    while exp < m:     # repeat the random experiments/assignments for m times
        # randomly assign birth date for n people
        months = [mon for mon in range(1,13)]
        birthdays = {}
        for people in range(0,n):
            month = random.choice(months)
            if month in [1,3,5,7,8,10,12]:
                day = random.choice([d for d in range(1,32)])
            elif month == 2:
                day = random.choice([d for d in range(1,30)])
            else:
                day = random.choice([d for d in range(1,31)])
            birthdays[people] = [month, day]          
        #print(birthdays)
        
        # if two same data, count += 1
        a = []
        for birthday in birthdays.values():
            if birthday not in a:
                a.append(birthday)
            else:
                count += 1
                break

        # start another assignment/experiment
        exp += 1
        
    #print(count)    
    # probability p = count / m 
    p = count / m
    return p
        
testparadox(10,5)   # --> p ~ 0.0
testparadox(10,10)  # --> p ~ 0.1
testparadox(10,10)  # --> p ~ 0
testparadox(20,10)  # --> p ~ 0.2
testparadox(10,20)  # --> p ~ 0.4
testparadox(10,23)  # --> p ~ 0.4
testparadox(10,25)  # --> p ~ 0.7
testparadox(10,40)  #  p ~ 0.7 ~ 1.0
testparadox(10,100)  # p = 1.0

```

```python
"""
Write a Python program that inputs a list of words, separated by whitespace, 
and outputs how many times each word appears in the list. You need not worry about efficiency at this point, 
however, as this topic is something that will be addressed later in this book.
"""
def wordlist(s):
    """
    input: s --> string
    output: words, numbers
    """
    dicts = {}
    s_list = [w.lower() for w in s.split()]
    for word in s_list:
        if word not in dicts:
            dicts[word] = 1
        else:
            dicts[word] += 1
    for word, count in dicts.items():
        print(word,': ', count)

s = 'Write a Python program that inputs a list of words'
wordlist(s)
            
    
```
