constant function: f(n) = 1
linear function: f(n) = n
logarithm function: f(n) = log(n), base is 2
n-log-n function: f(n) = nlog(n)
quadratic function: f(n) = n^2
summation function(d=1): f(n) = n(n+1)/2 = n^2/2 + n/2      f(n) = n*a1 + n(n-1)*d
geometric summation (q=2): f(n) = 2^n - 1                  f(n) = [q^(n+1) - 1] / (q - 1) ] 
cubic function: f(n) = n^3
polynomial function: f(n) = a0 + a1*n + a2*n^2 + ...
exponential function: f(n) = b^n

* Big-O notation:
f(n) <= cg(n) for n>n0  ==> f(n) is O(g(n)) 

select the highest growth rate as g(n):
1 < log(n) < n < nlog(n) < n^2 < n^3 =< a^n   for n>2 and a>=2

eg. f(n) = n^2 + nlogn < n^2 + n^2 = 2*n^2   ==> f(n) is O(n^2)

* Big-Omega notation:
f(n) >= cg(n), for n >= n0
eg. f(n) = n^2 + nlogn > nlogn + nlogn = 2*nlogn for n>=1 ==> f(n) is Ω(nlogn)
f(n) = 3nlogn - 2n = nlogn + 2n(logn - 1) >= nlogn for n>=2  ==> f(n) is Ω(nlogn)

* Big-Theta notation:
c1*g(n) <= f(n) <= c2*g(n), for n>=n0
eg. f(n) = 3nlogn + 4n + 5logn
   3*nlogn <= f(n) <= 12*nlogn  for n >= 1
   ==> f(n) is Θ(nlogn)
