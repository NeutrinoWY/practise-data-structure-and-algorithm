```python
s1 = {1,2,3,4}
s2 = {4,5,6}
sand = s1&s2  # belong to s1 and s2
sor = s1 | s2 # belong to s1 or s2
print(sand, sor)
s12 = s1 - s2 # belong to s1 but not belong to s2
s3 = s1 ^ s2  # belong to s1 or s2, but not belong to s1&s2
print(s12,s3)

print(s1 - sand)
print( s1 - s2 == s1 - s1&s2) 
# although the return values are the same, but they are not equal, they stored under different identifiers. 

print(sor - sand)
print(s1 ^ s2 == s1 | s2 - s1&s2)
```

```python
alpha = [1,2,3]
beta = alpha    # beta is an alias of alpha, they proxy for the same list at the memory
beta += [4,5]   # append [4,5] to the original alpha / beta
beta = beta + [6,7] # assign beta + [6,7] to a new variable beta, this beta points to another list at the memory
print(alpha)
```

```python
def fibonacci():    # define a generator, not a normal function
    a = 0
    b = 1
    while True:
        yield a     # using 'yield' to creat a generator
        future = a + b
        a = b
        b = future
```

```python
# simultaneous assignment, good mean to swap variables

def fibonacci():
    a,b = 0,1
    while True:
        yield a
        a,b = b, a+b
```

```python
# comprehension syntax [f(x) for x in iterator if conditions]
square = [k*k for k in range(1,100) if k % 2 == 0]
print(square)
# here is the list comprehension, it can also be tuple(), dict {}, set {}
```
