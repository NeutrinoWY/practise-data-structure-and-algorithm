cell of array array[index i] each cell should have the same bits of storage. 

compact array:  the array is contiguously storing the bits that represent the primary data 
word = 'sample' , array with 8*2*6 = 96 bits, one character for 2 bytes, 1 byte = 8 bits

string = 'hello world', cannot use compact array, because the length of each element is different, but we need same length of cell for each element. ---> referential structure. 
referential structural array: the array is contiguously storing the bits that represent the reference to the primary data
string = 'hello world'. 
string --> the address of start byte, for instance byte#1234
string[0] --> the 64 bits from 1234 to 1297 stores the reference(address byte#5100) of the value of string[0]
byte#5100 to byte#5110 stores the value 'hello'

compare to referential structure:
(1) compact array needs less memory (referential structure needs an extra memory for storing the references)
(2) primary data are stored consecutively in compact array --> high performance computing, quicker than referential structure in which need to read reference first then read value

The storage of python's list: 
Python's list is a dynamic array.
When user define a list with 5 elements, python will reserve more memory for the list. 
If user keep append the list, the old reserved memory space would be exhausted. Python would initiate a new list with 
more memory, and its prefix is as the old list. Then the system relaim the memory of the old list. 



```python
import sys
word = 'sample'   # word <-> address <-> value 
s = 'hello world'  # s <-> start address of references <-> reference <-> address of value <-> value

# the whole memory of word, the whole memory of s
a = sys.getsizeof(word)
b = sys.getsizeof(s)
print(a,'\n', b)
```

```python
class scoreboard:
    
    def __init__(self, capacity=10):
        self._capacity = capacity
        self._board = [None]*capacity
        self._n = 0
    
    def __getitem__(self,k):
        return self._board[k]
    
    def __str__(self):
        return '\n'.join(str(self._board[j]) for j in range(self._n))

    def add(self,entry):
        score = entry.get_score()
        good = self._n < len(self._board) or score > self._board[-1].get_score()
        if good:
            if self._n < len(self._board):
                self._n += 1
                
            j = self._n - 1    # the last index of scoreboard
            while j > 0 and self._board[j-1].get_score < score:
                self._board[j] = self._board[j-1]  # shift the score backward
                j -= 1   # index goes forward
                self._board[j] = score
                
        
    
```

```python
class Cipher:
    
    def __init__(self,shift):
        encode = [None]*26
        decode = [None]*26
        for k in range(26):
            encode[k] = chr((k+shift)%26+ord('A'))
            decode[k] = chr((k-shift)%26+ord('A'))
        self._forward = ''.join(encode)
        self._backward = ''.join(decode)
        
    def _transform(self, original, code):
        msg = list(original)
        for k in range(len(msg)):
            if msg[k].isupper():
                j = ord(msg[k])-ord("A")
                msg[k]=code[j]
        return ''.join(msg)
    
    def encrypt(self, message):
        return self._transform(message, self._forward)
    
    def decrypt(self, secret):
        return self._transform(secret, self._backward)
    
if __name__ == "__main__":
    cipher = Cipher(3)
    message = "THE EAGLE IS IN PLAY; MEET AT JOE'S."
    coded = cipher.encrypt(message)
    print(coded)
    answer = cipher.decrypt(coded)
    print(answer)
    
```

```python
class TicTacToe:
    def __init__(self):
        self._board = [[' ']*3 for i in range(3)]
        self._player = 'X'
    
    def mark(self, i, j): 
        if not 0<=i<=2 and 0<=j<=2:
            raise IndexError('Invalid board position')
        if self._board[i][j] != ' ':
            raise ValueError('Board position occupied')
        if self.winner() is not None:
            raise ValueError('Game is already completed')
        self._board[i][j] = self._player
        if self._player == 'X':
            self._player = 'O'
        else:
            self._player = 'X'

        
    def _is_win(self, mark):
        board = self._board
        return (mark == board[0][0] == board[0][1] == board[0][2] or
               mark == board[0][0]==board[1][1] == board[2][2] or 
               mark == board[0][0]==board[1][0] == board[2][0] or
               mark == board[0][1]==board[1][1] == board[2][1] or
               mark == board[0][2]==board[1][2] == board[2][2] or
               mark == board[1][0]==board[1][1] == board[1][2] or
               mark == board[2][0]==board[2][1] == board[2][2])
    
    def winner(self):
        for mark in 'XO':
            if self._is_win(mark):
                return mark
            return None
    
    def __str__(self):
        rows = ['|'.join(self._board[r]) for r in range(3)]
        return "\n-----\n".join(rows)
        
        
game = TicTacToe()
str(game)
        
```
