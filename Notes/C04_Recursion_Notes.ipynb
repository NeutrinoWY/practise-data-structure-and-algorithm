```python
"""IMPORTANT"""
"""IMPORTANT"""
"""IMPORTANT"""
# binary search, check whether a given target is in a given list. O(logn)
def bsearch(ls,target,left, right):
    """ls is a sorted list, that is the premise to use binary search algorithm"""
    if left > right:
        return False    # the situation that target is not in ls is considered here
    else:
        mid = (left + right) // 2
        if target == ls[mid]:
            return mid
        elif target < ls[mid]:
            bsearch(ls,target,left,mid - 1)
        else:
            bsearch(ls,target,mid+1,right)       # tail recursion
        
        
ls = [2,3,5,7,8,10]
bsearch(ls,7,0,len(ls))
```

```python
# key elements of recursion: function(N), function(N-1) and restrict conditions
# factorial n! = n*(n-1)*(n-2)*...*2*1
def factorial(n):
    if n == 1: 
        return 1
    elif n > 1:
        return n * factorial(n-1)          # non-tail recursion
    else:
        raise ValueError ('parameter should be positve integer. ')
  
factorial(5)
```

```python
# drawing ruler

def drawline(tick_length, tick_label=''):
    """draw one line with given tick length"""
    line = '-' * tick_length
    if tick_label:
        line += ' ' + tick_label
    print(line)

def drawinterval(central_length):
    """draw central line based on a central tick length"""
    if central_length > 0:
        drawinterval(central_length - 1)
        drawline(central_length)
        drawinterval(central_length - 1)
    
def drawruler(num_cm, major_length):
    """draw ruler with given number of centimeter and major tick length"""
    drawline(major_length,'0')
    for j in range(1, num_cm + 1):      # finish drawing one unit in one loop
        drawinterval(major_length - 1)
        drawline(major_length, str(j))

drawruler(3,3)
```

```python
# sequential search, O(n). 
# When the given list is not sorted, we use sequential search to check the element one by one
def ssearch(ls,target):
    i = 0
    while i < len(ls):
        if ls[i] == target:
            return i
        i += 1
    return False

ls = [2,3,5,7,8,10]
ssearch(ls,5)
   
```

```python
# calculate the usage of a path, O(n^2)，multiple recursion
import os
def disk_usage(path):
    total = os.path.getsize(path)  # the usage of the directory
    if os.path.isdir(path):
        for item in os.listdir(path):
            sub_path = os.path.join(path,item)
            total += disk_usage(sub_path)
    print('{0:<8}'.format(total), path)     
    return total

path = "/Users/yunwang/Downloads/Algorithms/Lecture"
disk_usage(path)
    
```

```python
# computing power with recursive algorithm, O(n)
def power(a,n):
    if n == 0:
        return 1
    return a * power(a,n-1)

power(2,5)
```

```python
# computing power with recursive algorithm, O(logn)
def power(a,n):
    if n == 0:
        return 1
    else:
        partial = power(a, n//2)          # since the obvious symetric, it's natural to reduce the process this way
        result = partial * partial
    if n % 2 == 1:
        result *= a
    return result

power(2,5)
```

```python
# reversing a squence with recursion, O(logn)
# linear recursion: function makes one recursion call
def reverse(s, start, stop):
    if start < stop - 1:
        s[start], s[stop-1] = s[stop-1], s[start]
        reverse(s, start+1, stop-1)       # tail recursion
    return s

s = [1,3,5,7,9]
reverse(s,0,len(s))
```

```python
# using recursion to caculate sum of a given list, running time:O(n), memory space: O(n)
def linear_sum(S,n):
    if n == 0:
        return 0
    else:
        return linear_sum(S,n-1) + S[n-1]         
# note: if it's a function call plus other, should add return in front of it
# if it's just a function call, don't use return in front of it
         
S = [1,2,3,4,5]
linear_sum(S,5)
```

```python
# function makes two recursion call -- binary recursion
# running time f(n) = 2n -1, O(n), memory space O(logn)
def binary_sum(S,start,stop):
    if start >= stop:
        return 0
    elif start == stop - 1:
        return S[start]
    else: 
        mid = (stop + start) // 2
        return binary_sum(S,start,mid) + binary_sum(S,mid,stop)
    
S = [1,2,3,4,5]
binary_sum(S,0,len(S))
```

```python
from itertools import permutations
def portfolio(seq):
    perm = permutations(seq)
    count = 0
    for item in perm:
        s = ''
        for i in item:
            s += i
        print(s)
        count += 1
    print(count)

portfolio(['a','b','c','d'])
```
