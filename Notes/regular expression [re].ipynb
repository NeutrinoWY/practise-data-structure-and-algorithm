```python
import re
x = "he said: here you go!!"
y = re.findall('^h.+!', x) #^h.(the string start with h), + (one or more until the last)
z = re.findall('^h.+?!',x) # +? (one or more until the nearest)
print(y)
print(z)
```

```python
log = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
email = re.findall('\S+@\S+',log)  # \S+ (non-whitespace character)
print(email)

atpos = log.find('@')
sppos = log.find(' ', atpos)  # find the position of whitespace, starting from atpos
domain = log[atpos+1:sppos]
print(domain)
```

```python
# another way to find domain
email1 = log.split()[1]
domain1 = email1.split('@')[1]
print(domain1)
```

```python
domain2 = re.findall('^From .*@([^ ]+)', log)
print(domain2)
```

^From - start from 'From'
.*  -  skip bunch of characters
@   -  until @
(   -  begin extracting
[^ ] -   extract non whitespace character
+    -   extract many 
)    -   end of extracting
