```python
"""
Describe a recursive algorithm for finding the maximum element in a se- quence, S, of n elements. 
What is your running time and space usage?
"""

def search_max(s,n):
    if n == 1:
        return s[0]
    return max(s[n-1], search_max(s, n-1))

if __name__ == '__main__':
    s = [13,5,6,32,68,35,6,7]
    n = len(s)
    print(search_max(s,n))
```
