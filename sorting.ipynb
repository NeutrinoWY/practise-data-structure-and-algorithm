**Quick Sort**

```
# quick sort，先整体再局部，pivot，小于的去左边，大于pivot的去右边
def sort_intergers(a):
  if a == None or len(a) == 0:
    return a
  
  quick_sort(a, 0, len(a)-1)

# 递归的定义
def quick_sort(a, start, end):
  # 递归的出口
  if start >= end:
    return
  
  # 递归的拆解，partition分区，双指针
  # 左边都小于pivot，右边都大于pivot
  left, right = start, end
  # 可以随机选取，一般来说，选取中间的点作为pivot
  pivot = a[start + (end - start)//2]   #不用(start+end)//2，以防数值过大出现外溢

  # while循环在left>right时结束
  while left <= right:
    # 找左边开始找第一个不小于pivot的值的位置
    while left <= right and a[left] < pivot:
      left += 1
    # 从右边开始找第一个不大于pivot的值的位置
    while left <= right and a[right] > pivot:
      right -= 1
    
    # 此时 a[left] >= pivot and a[right] <= pivot
    # 如果此时还满足left在right左边，则交换两个值，这样小于等于的就去了pivot左边，大于等于的就去了pivot右边
    if left <= right:
      a[left], a[right] = a[right], a[left]
      left += 1
      right -= 1
    
  quick_sort(a, start, right)
  quick_sort(a, left, end)

A = [3,2,41,3,56,34]
sort_intergers(A)
A
```

**Merge Sort**

```
# 先局部再整体，不断二分直到一个数值是一组，然后将二分的按照顺序merge成一组
def sort_intergers(A):
  if A == None or len(A) == 0:
    return A
  
  # 新开一个tmp数组用来存储按次序merge的数
  tmp = [0 for _ in range(len(A))]
  merge_sort(A, tmp, 0, len(A) - 1)

# 递归定义
def merge_sort(A, tmp, start, end):
  # 递归出口
  if start >= end:
    return
  
  # 递归的拆解
  # 取中间位置把数组分成两段
  mid = start + (end - start) // 2

  # 左右两边再各自divide
  merge_sort(A, tmp, start, mid)
  merge_sort(A, tmp, mid+1, end)

  # 按照大小排序合并左右两边
  merge(A, tmp, start, mid, end)

# 利用双指针比较左右两个数组的大小，排序写进tmp中
def merge(A, tmp, start, mid, end):
  # [start, mid], [mid+1, end] 左右两区
  # 用两个index记录左右两边分别处理到哪里
  left, right = start, mid + 1

  # 在tmp中对应的位置
  index = start

  while left <= mid and right <= end:
      # 左小于右
    if A[left] <= A[right]:
      tmp[index] = A[left]
      index += 1
      left += 1
    else:
      tmp[index] = A[right]
      index += 1
      right += 1
    
    # 左边数组还有剩下的
  while left <= mid:
    tmp[index] = A[left]
    index += 1
    left += 1

    # 右边数组还有剩下的
  while right <= end:
    tmp[index] = A[right]
    index += 1
    right += 1
  
  # 更新原数组
  for i in range(start, end+1):
    A[i] = tmp[i]

A = [3,2,41,3,56,34]
sort_intergers(A)
A
```

思路：用set或者dictionary来记录每个元素出现的次数，然后修改原数组就可以了

```
# use dictionary
def sort_color(colors):
  if len(colors) == 0 or colors == None:
    return colors
  
  d = {}
  for c in colors:
    if c not in d:
      d[c] = 1
    else:
      d[c] += 1
  # 也可以用set，因为这里的元素种类是固定的
  
  colors[0:d[0]] = [0]*d[0]
  colors[d[0]:d[0]+d[1]] = [1]*d[1]
  colors[d[0]+d[1]:] = [2]*d[2]

  return colors

sort_color([0,1,0,2,1])
```

#532. **Reverse Pairs**

input: [4,1,3,2,5]；   
output: 4；    
数组中有(4,1),(4,3),(4,2),(3,2)这四个逆序对    
要求：时间复杂度为O(nlogn)

思路一：嵌套的for循环 O(n^2)     
思路二：时间复杂度为O(nlogn)的涉及到数组元素大小次序的算法有quick sort和merge sort； quick sort是先整体再局部排序，用pivot来做，使左边元素小于pivot，右边元素大于pivot，在这题目中不适用；merge sort是先局部再整体，将数组元素二分再二分，直到一个元素为一组，然后比较左右组的元素大小再merge，这个能用在本题中；只要在merge 比较元素大小时，如果右边元素小于左边元素，那么就可以记录一个逆序对。

```
def reverse_pairs(A):

  if A == None or len(A) == 0:
    return 0
  
  tmp = [0 for _ in range(len(A))]
  return merge_sort(A, tmp, 0, len(A)-1)

def merge_sort(A, tmp, start, end):
  if start >= end:
    return 0
  
  mid = start + (end - start) // 2
  # 记录逆序对个数
  sum = 0
  # 左区间右区间做归并排序
  sum += merge_sort(A, tmp, start, mid)
  sum += merge_sort(A, tmp, mid+1, end)

  # 合并两个有序数组
  sum += merge(A, tmp, start, mid, end)
  return sum

  # 合并两个有序数组
def merge(A, tmp, start, mid, end):
  left, right = start, mid+1
  index = start
  sum = 0
  while left <= mid and right <= end:
    if A[left] <= A[right]:
      tmp[index] = A[left]
      index += 1
      left += 1
    else:
      tmp[index] = A[right]
      # 记录reverse pairs个数最重要一步
      sum += mid - left + 1    # left指针后面的数都比A[right]大
      index += 1
      right += 1 
  
  # 左边的数组还剩下部分
  while left <= mid:
    tmp[index] = A[left]
    index += 1
    left += 1
  
  # 右边的数组还剩下部分
  while right <= end:
    tmp[index] = A[right]
    index += 1
    right += 1

  # 更新数组
  for i in range(start, end+1):
    A[i] = tmp[i]
  
  return sum


reverse_pairs([4,1,3,2,5])
```
