# 35. **【翻转链表】**

输入: 1->2->3->null
输出: 3->2->1->null

```
def reverseLinkedList(head):
  prev = None
  while head != None:
    temp = head.next   #将要翻转节点的下一个节点记录下来，以免丢失
    head.next = prev   #翻转
    prev = head        # prev后移一个节点
    head = temp        # head也后移一下节点
  return prev     #注意是return prev，head最后会移到null上去
```

# 225.**【寻找节点】**
在链表中找值为 value 的节点，如果没有的话，返回空(null)。

输入:  1->2->3 and value = 3；  
输出: 最后一个结点

如果有多个相同的结点，返回找到的第一个结点

从头开始遍历链表，遇到等于value的值，返回该node，若没有这个值则返回None。

```
def searchNode(head, value):
  if head == None: return None

  while head != None:
    if head.val == value:
      return head
    head = head.next
    return None
```

返回所有的相同节点位置

```
def searchNode(head, value):
  if head == None: return None

  record = []
  count = 0
  while head != None:
    if head.val == value:
      record.append(count)
    count += 1
    head = head.next
  return record

```

# 228.**【寻找中点】**
找链表的中点，并返回这个节点

输入:  1->2；  
输出: 1	；  
样例解释: 如果长度是偶数，则返回中间偏左的节点。

思路： 用两个指针，一个快指针，一个慢指针，快指针每次移动节点数为慢指针的1倍，这样当快指针达到链表末尾时，慢指针就在中间位置。  
奇数节点链表：快指针在最后一个node时，慢指针在中间；  
偶数节点链表：快指针停止不能移动时，慢指针在中间偏左

```
def searchMiddleNode(head):
  if head == None: return None

  fast = head
  slow = head
  while fast.next != None and fast.next.next != None:
    fast = fast.next.next
    slow = slow.next
  return slow

```

# 219.**【插入节点】**
在排序链表中插入一个节点

输入：head = 1->4->6->8->null, val = 5；   
输出：1->4->5->6->8->null

思路：用一个指针遍历链表，如果遇到的node值 < val, 则往后移动指针；当指针后一个节点值大于val或者为None时，停止移动，插入新节点。   
技巧：用一个dummy节点放在head前面，这样可以处理head==None的情况，同时最后可以直接返回dummy.next

```
def insertNode(head, value):

  dummy = ListNode(-1,head)
  currNode = dummy

  while currNode.next != None and currNode.next.val < value:
    currNode = currNode.next
  
  insert_node = ListNode(value, currNode.next)
  currNode.next = insert_node

  return dummy.next

```

# 165.**【合并列表】**
将两个排序链表合并为一个新的排序链表

样例2:

	输入:  list1 =  1->3->8->11->15->null, list2 = 2->null

	输出: 1->2->3->8->11->15->null

方法1： 将list2一个个的加入大list1中 （要判断的情况比较复杂）

时间复杂度O(N), N=len(s1)+len(s2); 空间复杂度O(1)     

test: (1) head1==None, head2 != None; (2) head1 != None, head2 == None;  (3) head1 != None, head2 != None

```
def mergeList(head1, head2):
  # dummy2的好处是不用特殊考虑head2为None的情况
  dummy1 = ListNode(-1, head1)
  dummy2 = ListNode(-1, head2)
  curr1 = dummy1
  curr2 = dummy2

  if head1 == None: return head2

  while curr2.next != None:
    # 移动curr1指针直到其next值>= curr2.next；加上curr1.next判断条件，即如已是list1末尾则不再移动curr1指针
    while curr1.next != None and curr1.next.val < curr2.next.val:    #这里没有包含head1为None的特殊情况
      curr1 = curr1.next

    # 在list1中插入list2中指针curr2后的节点
    insert_node = ListNode(curr2.next.val)
    insert_node.next = curr1.next
    curr1.next = insert_node

    # 将两个指针都后移，开始插入新的节点
    curr1 = curr1.next
    curr2 = curr2.next
  
  return dummy1.next
```

方法二：生成一个新的链表

时间复杂度O(N), N = m+n; 空间复杂度O(m+n)

```
def mergeLists(head1, head2):
  dummy = ListNode(-1)
  currNode = dummy

  while head1 != None and head2 != None:
    if head1.val > head2.val:
      currNode.next = head1
      currNode = currNode.next
      head1 = head1.next
    else:
      currNode.next = head2
      currNode = currNode.next
      head2 = head2.next
  
  if head1 != None:
    currNode.next = head1
  else:
    currNode.next = head2
  
  return dummy.next
```

# 452.**【删除节点】**
删除链表中等于给定值 val 的所有节点

考虑val在链表前、尾、中三种情况，并且要考虑到连续都是val的节点情况;   
test: 3->3->2->1->3->3->1->2->3->null

时间复杂度O(n),空间复杂度O(1)

```
def deleteNodes(head, val):

  dummy = ListNode(-1, head)
  currNode = dummy 

  while currNode.next != None:
    # 如果next节点值=val,remove这个节点
    if currNode.next.val == val:
      currNode.next = currNode.next.next
    # 如果next节点！= val, 将currNode后移
    else:
      currNode = currNode.next

  return dummy.next
```

# 174. **【删除倒数nth节点】**
给定一个链表，删除链表中倒数第n个节点，返回链表的头节点; 链表中的节点个数大于等于n.   
Input:  list = 5->4->3->2->1->null, n = 2;   
Output: 5->4->3->1->null


思路： 用两个指针，一个走到链表末尾时，另一个指针正好走到要删除的倒数n节点那；
就是说，两指针开始就相距n节点，然后一起一个一个节点往后移。

时间复杂度O(n), 空间复杂度O(1)

```
def removeNthFromEnd(head, n):

  dummy = ListNode(-1, head)
  ahead = dummy
  behind = dummy
  # 先将一个指针移n节点，使得两个指针距离n节点
  for i in range(n):
    ahead = ahead.next
  # 再一起移动两个指针，直到前面的一个达到链表末尾，此时另一个指针的next(behind.next)就是倒数n节点位置
  while ahead.next != None:
    ahead = ahead.next
    behind = behind.next
  # 删除该节点
  behind.next = behind.next.next

  return dummy.next
  
```

#170.**【旋转链表】**
给定一个链表，旋转链表，使得每个节点向右移动k个位置，其中k是一个非负数

输入：1->2->3->4->5  k = 2；  
输出：4->5->1->2->3

思路：将后面k个节点整体移到链表前面；怎么找截断的地方？用两个指针，一个指针移动到最后一个节点时，另一个指针在截断节点的前面一个节点处。以上面的例子，一个指针到5时，另一个指针在3处，这样3后面就开始截断，整个移到链表前面。
(1) head = None;
(2) k > length(list);
(3) k < length（list);

(2)(3)可以整合成 k = k % length(list)


时间复杂度O(n), 空间复杂度O(1)

```
def getLength(head):     
  count = 1
  curr = head
  while curr.next != None:         #注意这里尽量不要移动head来数节点个数，因为如果将这段代码写在主代码中，会改变head的节点位置，从而影响整个主代码
    curr = curr.next
    count += 1
  return count


def rotateList(head, k):
  if head == None: return None

  n = k % getLength(head)
  behind = head
  ahead = head

  for i in range(n):
    ahead = ahead.next
  
  while ahead.next != None:
    ahead = ahead.next
    behind = behind.next
  
  ahead.next = head
  dummy = ListNode(-1,behind.next)
  behind.next = None

  return dummy.next

```
