**856. 句子相似性**

输入: words1 = ["fine","skills","acting"],       
words2 = ["fine","drama","talent"]，  
pairs = [["great","fine"],["drama","acting"],["skills","talent"]]

输出: false

解释:
"fine"和"fine"相同
"skills"和"drama"不相似
"acting"和"talent"不相似


```python
def is_similar_sentence(words1, words2, pairs):
  if (not len(words1) == len(words2)):    
    return False
  
  mp = {}
  for i in range(len(pairs)):
    S = set()
    if not mp.get(pairs[i][0]) == None:
      S = mp[pairs[i][0]]    # mp的value是一个set
  
    S.add(pairs[i][1])   # 将一对pair的第二个值加入到set中
    
    mp[pairs[i][0]] = S    # 将一对pair的第一个值作为key，第二个值的set作为value加入到mp字典中
  print(mp)

    

  for i in range(len(words1)):
    if words1[i] == words2[i]:
      continue
    # 如果mp中words1[i]作为key，对应的值是None， 或者相应的words2[i]不是mp[words1[i]]的值；......
    if (mp.get(words1[i]) == None or words2[i] not in mp[words1[i]]) and (mp.get(words2[i]) == None or words1[i] not in mp[words2[i]]):
      return False
  return True

if __name__ == "__main__":
  words1 = ["fine", "acting","skills"]
  words2 = ["fine","drama","talent"]
  pairs = [["great","fine"],["drama","acting"],["skills","talent"]]
  print(is_similar_sentence(words1, words2, pairs))
```

**644. 镜像数字**

一个镜像数字是指一个数字旋转180度以后和原来一样(倒着看)。例如，数字"69"，"88"，和"818"都是镜像数字。
写下一个函数来判断是否这个数字是镜像的。数字用字符串来表示。

思路：数字一共只有10个，分别列出一对对成镜像的数字对，放入dictionary中；
mirror_pairs = {'6':'9', '9':'6', '8':'8', '1':'1', '0':'0'}

偶数个数字，前后一对对数字应该属于mirror_pairs中的一对；奇数个数字，前后一对对应该属于mirror_paris中的一对，中间一个数字应该in mirror_pairs.

```python
def is_mirror_number(s):
  if s is None or len(s) == 0:
    return True
  
  mirror_pairs = {'6':'9', '9':'6', '8':'8', '1':'1', '0':'0'}
  n = len(s)
  start = 0
  end = n - 1
  while start <= end:
    if s[start] not in mirror_pairs or mirror_pairs[s[start]] != s[end]:
      return False
    start += 1
    end -= 1
  return True

if __name__ == "__main__":
  S = ["69", "808", "44", ""]
  for s in S:
    print(is_mirror_number(s))
```

**521. 去除重复元素** 

给一个整数数组，去除重复的元素。

你应该做这些事

1.在原数组上操作     
2.将去除重复之后的元素放在数组的开头       
3.返回去除重复元素之后的元素个数

输入: nums = [1,3,1,4,4,2]            
输出: 4     
(原数组变为: [1,3,4,2,x,x])


```python
def deduplication(nums):
  if len(nums) == 0 or len(nums) == 1:
    return len(nums)
  
  if len(nums) == 2:
    if nums[0] == nums[1]:
      return 1
    else:
      return 2
  
  start = 0
  end = len(nums) - 1
  S = set()
  while start < end:
    if nums[start] not in S:
      S.add(nums[start])
      start += 1
    else: 
      if nums[end] not in S:
        nums[start], nums[end] = nums[end], nums[start]
      end -= 1
      
  return start

nums_list = [[1,3,1,4,4,2], [1,1,1,1,1,1], [2,1], [1], []]
for nums in nums_list:
  print(deduplication(nums))
```

**1035. 森林中的兔子**

在一个森林中，每个兔子都有一种颜色。兔子中的一部分（也可能是全部）会告诉你有多少兔子和它们有同样的颜色。这些答案被放在了一个数组中。

返回森林中兔子可能的最少的数量.

输入: [1, 1, 2]
输出: 5

思路： 分解各种情况如下       
[1,1] -> 2只兔子互相指认；     
[3,3,3,3] -> 4只兔子互相指认；        
[2] -> 1只兔子指认其他2只，该颜色一共3只兔子；                        
[2,2,2,2,2] -> 分为[2,2,2],[2,2] -> 前后对应两种颜色，各有3只，一共6只；  

总结：          
相同数字的个数小于等于该数字的部分属于同一种颜色的兔子，count_num <= num+1, -> num + 1; 数字个数超过num+1的部分组成新的颜色组


```python
def leastNum(answers):
  if answers is None or len(answers) == 0:
    return 0
  
  # 现将数字及对应的个数记录下来
  # {number: count of this number}
  mps = {}
  for num in answers:
    if num in mps:
      mps[num] += 1
    else:
      mps[num] = 1
  print(mps)
  
  # unique numbers
  num_set = mps.keys()
  print(num_set)

  result = 0
  for num in num_set:
    # 整除，属于不同颜色组的组数
    n1 = mps[num] // (num + 1)
    # 余数，剩下来的属于一个颜色组
    if mps[num] % (num + 1) == 0:
      n2 = 0
    else:
      n2 = 1
    
    result += (n1 + n2) * (num + 1)
  
  return result

if __name__ == "__main__":
  ans = [2,2,2,2,2,1,3]
  print(leastNum(ans))
```
