一个有效的路径，指的是从根节点到叶节点的路径。

input    
binary tree: {1,2,4,2,3}               
target sum: 5  

output: [[1, 2, 2],[1, 4]]

思路：涉及到深度路径的计算，那么可以用depth first search. 分成左右两个部分，分治法用递归实现，在遍历树过程中记录路径之和。left_sum = curr_node.val + curr_node.left.val, right_sum = curr_node.val + curr_node.right.val，

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left = None
    self.right = None

class Solution:
  def binaryTreePathSum(self, root, target):
    result = []
    path = []
    self.dfs(root, path, result, 0, target)
    return result

  # 递归的定义
  def dfs(self, root, path, result, sum, target):
    # 特殊情况
    if root is None:
      return
    
    path.append(root.val)
    sum += root.val
    # 递归的出口
    if root.left is None and root.right is None and sum == target:
      result.append(path[:])
    
    # 递归的分解
    self.dfs(root.left, path, result, sum, target)
    self.dfs(root.right, path, result, sum, target)

    # 回到父节点
    path.pop()


```

```
from collections import deque
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left, self.right = None, None

class Solution:
  def maxDepth(self, root):
    if root is None: return 0
    queue = deque([root])
    result = 0
    while queue:
      for _ in range(len(queue)):
        node = queue.popleft()     # 将节点pop出来
        if node.left:              # 如果该节点有子节点，则push进去
          queue.append(node.left)
        if node.right:
          queue.append(node.right)
      result += 1                  # 该节点的层数
    return result

if __name__ == "__main__":
  root = TreeNode(1)
  leftNode = TreeNode(2)
  rightNode = TreeNode(3)
  root.left = leftNode
  root.right = rightNode
  leftNode.left = TreeNode(4)
  leftNode.right = TreeNode(5)
  rightNode.right = TreeNode(6)

  S = Solution()
  print(S.maxDepth(root))
```

#67. **二叉树的中序遍历**

思路：中序遍历，先左再节点再右。先解决两层二叉树的遍历，用递归的方法扩展到更多层的二叉树（分治法）

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left, self.right = None, None

class Solution:
  def inorder_tranversal(self, root):
    if root is None:
      return []
    
    dummy = TreeNode(0)
    dummy.right = root
    stack = [dummy]

    inorder = []

    while stack:
      node = stack.pop()
      if node.right:
        node = node.right 
        while node:
          stack.append(node)
          node = node.left
      if stack:
        inorder.append(stack[-1].val)
    
    return inorder



if __name__ == "__main__":
  root = TreeNode(1)
  leftNode = TreeNode(2)
  rightNode = TreeNode(3)
  root.left = leftNode
  root.right = rightNode
  leftNode.left = TreeNode(4)
  leftNode.right = TreeNode(5)
  rightNode.right = TreeNode(6)

  S = Solution()
  print(S.inorder_tranversal(root))
```

#**177. 把排序数组转换为高度最小的二叉搜索树**

输入：[1,2,3,4,5,6,7]      
输出： {4,2,6,1,3,5,7}

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left, self.right = None, None

class Solution:
  def sorted_array_to_bst(self, A):
    return self.convert(A, 0, len(A)-1)

  def convert(self, A, start, end):
    # None or empty
    if start > end:
      return None
    # with only one element
    if start == end:
      return TreeNode(A[start])
    
    mid = (start + end) // 2
    root = TreeNode(A[mid])
    root.left = self.convert(A, start, mid - 1)
    root.right = self.convert(A, mid + 1, end)

    return root

    
```

#93. **平衡二叉树**

给定一个二叉树,确定它是高度平衡的        
一棵高度平衡的二叉树的定义是：一棵二叉树中每个节点的两个子树的深度相差不会超过1

```
# 在每一步计算左右子树高度时，顺便判断高度差是否小于1
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left, self.right = None, None

class Solution:
  def is_balance_tree(self, root):
    balanced, _ = self.depth(root)
    return balanced

  def depth(self, root):
    # 递归出口
    if root is None: 
      return True, 0
    
    left_balanced, depth_left = self.depth(root.left)
    right_balanced, depth_right = self.depth(root.right)
    depth = max(depth_left, depth_right) + 1

    if left_balanced is False or right_balanced is False:
      return False, depth
    if abs(depth_left - depth_right) > 1:
      return False, depth

    return True, depth

```

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left, self.right = None, None

class Solution:
  def insert_node(self, root, node):
    if root is None:
      return node

    val = node.val
    curr_node = root
    while curr_node != node:       # while条件的设定
      if curr_node.val < val:
        if curr_node.right is None:
          curr_node.right = node
        curr_node = curr_node.right

      if curr_node.val > val:
        if curr_node.left is None:
          curr_node.left = node
        curr_node = curr_node.left
    
    return root


  def preorder_tranversal(self, root):
    if root is None:
      return []
    
    stack = [root]
    order = []

    while stack:
      node = stack.pop()
      order.append(node.val)
      # 将node.right放在node.left条件前面，这样stack.pop()出来的就先是left
      if node.right:             
        stack.append(node.right)
      if node.left:
        stack.append(node.left)
    return order


if __name__ == "__main__":

  root = TreeNode(3)
  root.left = TreeNode(2)
  root.right = TreeNode(5)
  node = TreeNode(1)

  s = Solution()
  s.insert_node(root, node)
  print(s.preorder_tranversal(root))

```

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left, self.right = None, None

class Solution:
  def search_node(self, root, value):
    # use pre-order tranversal
    if root is None:
      return None
    stack = [root]
    while stack:
      node = stack.pop()
      if node.val == value:
        return node

      if node.right:
        stack.append(node.right)
      if node.left:
        stack.append(node.left)
    return None
```

#1534. **将二叉搜索树转换成已排序的双向链表**

时间复杂度O(n), 空间复杂度O(n)

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left = None
    self.right = None
  
  def treeToDoubleList(self, root):
    if not root:
      return root
    head = None
    tail = None
    prev = None
    for v in self.inorder_tranverse(root):
      if head is None: head = v
      tail = v
      if prev is not None:
        prev.right = v
        v.left = prev
      prev = v
    head.left = tail
    tail.right = head
    return head 
    
  def inorder_tranverse(self, root):
    if not root:
      return
    left = root.left
    for node in self.inorder_tranverse(left):
      yield node
    yield root
    right = node.right
    for node in self.inorder_tranverse(right):
      yield node
  
```

#701. **Trim BST**

给定一个有根的二叉搜索树和两个数字min和max，修剪树使得所有的数字都是在min和max之间（包括min和max），这个所得的树仍然是合法的二叉搜索树.

思路：     
（1） 若根节点值小于最小值，则递归调用右子树并返回右子树；    
（2） 若根节点值大于最大值，则递归调用左子树并返回左子树；    
（3） 否则修剪左子树，右子树并返回根节点。      

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left = None
    self.right = None

def trim_bst(root, min, max):
  if root is None:
    return root
  if root.val < min:
    return self.trim_bst(root.right, min, max)
  elif root.val > max:
    return self.trim_bst(root.left, min, max)
  else:
    root.left = self.trim_bst(root.left, min, max)
    root.right = self.trim_bst(root.right, min, max)
  return root


```
