Rabbit in forest

```
def least_rabbit_num(answers):
  d = {}
  # record the numbers of different answers
  for num in answers:
    if num not in d:
      d[num] = 1
    else:
      d[num] += 1
  
  rabbit_num = 0
  for k, v in d.items():
    # 如果相同的回答只有1个，那么这属于一种颜色，这种颜色的兔子个数为k+1
    if v == 1:
      rabbit_num += k+1
    # 相同的回答超过1个,[2,2,2,2,2,2,2], 每3个2是一种颜色，最后一个2是一种颜色
    else:
      rabbit_num += v//(k+1) * (k+1) + v % (k+1) * (k+1)
  
  return rabbit_num
    

least_rabbit_num([1,2,1,2,2,3,2])
# 1,1->2只相同颜色的兔子；2，2，2，2 -> 前3个2是一种颜色兔子，有3只，后一个2是一种颜色兔子有3只； 3-> 4只另一种颜色兔子
# 2+3+3+4=12
```

Sentence similarity

```
# dictionary.setdefault(key, value)
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# 将"color":set()加入到car这里dictionary中，并且return value set()
x = car.setdefault("color", set())

print(x)
print(car)

# 在key对应的value set()中加入值
car["color"].add("red")
print(car)
```
