**求最大公约数**

用较大的数除以较小的数，再用除数除以出现的余数（第一余数），再用第一余数除以出现的余数（第二余数），如此反复，直到最后余数是0为止，最后的除数就是这两个数的最大公约数。

```python
def gcd(big, small):
    if small != 0:
        return gcd(small, big % small)
    else:
        return big

print(gcd(51, 34))
```

**65. 求两个排序数组的第k小的数**

在两个升序排序数组中，求他们合并到一起之后的第k小的数。     
时间复杂度要求：O(log(n+m))，其中 n, m 分别为两个数组的长度

思路： o(logN) -》 二分法，O(1)内就排除一半     
（1） 两个数组比较第k/2个数 （合并起来就是k），第k/2的数小的数组，前面的部分可以全部排除；    
（2）已经排除k/2， 那么要找的数就排在剩下来的第（k - k/2）的位置；    
（3）比较两个数组第k-k/2的数，谁小就排除，接着要找的数就排在第k - k/2 - (k-k/2）位；    
（4）依次类推下去直到k=1，就是比较两个数组排在第一位的数


```python
def find_kth_smallest_element(list1, list2, k):
  
  if k//2 == 1:
    return min(list1[0], list2[0])

  if list1[k//2 - 1] <= list2[k//2 - 1]:
    find_kth_smallest_element(list1[k//2:], list2, k - k//2)
  else:
    find_kth_smallest_element(list1, list2[k//2:], k - k//2)



list1 = [1,3,5,7]
list2 = [2,4,6,8,10]
print(find_kth_smallest_element(list1, list2, 4))

```

***461. Kth smallest element in two unsorted lists**

**分解质因数**

以 sqrt{n} 为时间复杂度的算法并不多见，最具代表性的就是分解质因数了。

具体步骤：         
1. 记up = sqrt{n}，作为质因数k的上界, 初始化k=2；    
2. 当k <= up 且 n不为1 时，执行步骤3，否则执行步骤4；       
3. 当n被k整除时，不断整除并覆盖n，同时结果中记录k，直到n不能整出k为止。之后k自增，执行步骤2；     
4. 当n不为1时，把n也加入结果当中，算法结束。     

几点解释：    
* 不需要判定k是否为质数，如果k不为质数，且能整出n时，n早被k的因数所除。故能整除n的k必是质数。       
* 为何引入up？为了优化性能。当k大于up时，k已不可能整除n，除非k是n自身。也即为何步骤4判断n是否为1，n不为1时必是比up大的质数。     
* 步骤2中，也判定n是否为1，这也是为了性能，当n已为1时，可早停。

复杂度分析：      
* 最坏时间复杂度O(sqrt (n) )。当n为质数时，取到其最坏时间复杂度。
* 空间复杂度O(log(n)), 当n质因数很多时，需要空间大，但总不会多于O(log(n))个

```python
def primeFactorization(n):
    result = []
    up = int(math.sqrt(n));
    
    k = 2
    while k <= up and n > 1: 
        while n % k == 0:
            n //= k
            result.append(k)
        k += 1
            
    if n > 1:
        result.append(n)
        
    return result
```

**分块检索算法**

给一个数组，数字范围在[0,10000], 求比当前数小的前面的数的个数。
eg. A = [1, 2, 4, 6, 5, 3], result = [0, 1, 2, 3, 3, 2]

分块检索：不是数字分块，而是数所在的区间分块；例如数字所在的区间范围是0-10000，可以将每100间隔分为一个小区间，那么就有101个区间：0-99，100-199,...,9900-9999，10000-10000

```python
class Block:
  def __init__(self):
    # Block的长度
    self.total = 0
    # Block中的{数：数的个数}
    self.counter = {}

class BlockArray:
  def __init__(self):
    # initialize blocks
    self.blocks = [Block() for _ in range(max_value // 100 + 1)]
  
  def count_smaller(self, value):
    """
    count the number of values that are located before and smaller than the target value.
    """
    count = 0
    block_index = value // 100
    for in in range(block_index):
      # value所在的block之前的blocks的总长度
      count += self.blocks[i].total 
    
    counter = self.blocks[block_index].counter
    for val in counter:
      if val < value:
        count += counter[val]
    return count
  
  def insert(self, value):
    block_index = value // 100
    block = self.blocks[block_index]
    block.total += 1
    block.counter[value] = block.counter.get(value, 0) + 1
  
class Solution:
  def count_smaller_number(self, A):
    """
    @param A: a integer list/array
    @return: a list of integers that state the number of smaller values before current index.
    """
    if not A:
      return []
    
    block_array = BlockArray(10000)
    results = []
    for a in A:
      count = block_array.count_smaller(a)
      result.append(count)
      block_array.insert(a)
    return results

```
