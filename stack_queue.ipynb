#541. **zigzag iterator**

vecs = [[1,2,3],[1,2],[1],[]] 
result = [1,1,1,2,2,3]

思路一： 用一个行指针和一个列指针，先移动行指针，遇到不为空值则加入到result中，遍历完行之后，列指针移动到下一列，再开始重复遍历行，最后result中就有按照zigzag路径得到的所有非空值。
note：这里要注意的是，当列指针>行长时，值为空; 所以有值的位置是 行长 > 列指针, 注意行为空时其行长为0，而列指针从0开始，所以条件是行长>列指针，而不能有等号=

iterator包含两个函数，一个是hasNext(),一个是next()，另外还有init。

将result定义为deque type，将非空值一个个加入到queue中，当调用next()时就提取最前面的一个值，hasNext()等同于判断queue是否为空。

时间复杂度O(n), 空间复杂度O(n), n-非空元素个数； 

```
from collections import deque
class zigzagIterator:

  def __init__(self,vecs):
    rows = len(vecs)
    self.vecs = vecs
    self.rows = rows 

    #列数
    columns = 0
    for i in range(rows):
      columns = max(len(self.vecs[i]),columns)

    result = deque()    #用来装按照zigzag的路径得到的非空值
    #遍历列
    col_index = 0
    while col_index < columns:
      #遍历行
      row_index = 0
      while row_index < rows:
        if len(vecs[row_index]) > col_index:        #key point: 需要将空行去掉，否则用indices取值会报错index out of range
          result.append(vecs[row_index][col_index])
        # 下移一行
        row_index += 1
      #后移一列
      col_index += 1

    self.queue = result
  
  def __len__(self):
    return len(self.queue)
  
  def hasNext(self):
    return len(self.queue) != 0  
  
  def next(self):
    if self.hasNext:
      return self.queue.popleft()
    else:
      print("The iterator is empty.")

    
  

if __name__ == "__main__":
  vecs = [[1,2,3],[1,2],[1],[]]
  iter1 = zigzagIterator(vecs)

  vecs = [[],[1],[],[1,2]]
  iter2 = zigzagIterator(vecs)

  for i in range(len(iter1)):
    print(iter1.next())
  
  for i in range(len(iter2)):
    print(iter2.hasNext())
    print(iter2.next())
  
```

思路二: 将每一行整体加入到queue中，每次pop一个first value，就将该行移至queue尾。 hasNext()行数判断queue中是否还有非空值，next()行数pop最前面一行最前面一个值。

```
class zigzagIterator:
  def __init__(self, vecs):

    # 将非空行作为iterator，与行长一起整体加入到queue中
    queue = deque()
    for i in range(len(vecs)):
      if len(vecs[i]) != 0: 
        queue.append((iter(vecs[i]), len(vecs[i])))       #这里用到一个技巧：直接将list转换成iterator

    self.vecs = vecs
    self.queue = queue

  def __len__(self):
    return len(self.queue)
  
  def next(self):
    # pop出第一行第一个值，如这行剩下的值不为空，则将其加入到对尾

    #pop第一行· deque.popleft（）
    row, length = self.queue.popleft()
    # pop第一个值
    val = next(row)
    length -= 1
    # 没有pop完则重新加入到队尾
    if length != 0:
      self.queue.append((row, length))
    return val

  def hasNext(self):
    # 判断queue中是否还有元素
    return len(self.queue) > 0


if __name__ == "__main__":
  vecs = [[1,2,3],[1,2],[1],[]]
  iter1 = zigzagIterator(vecs)

  vecs = [[],[1],[],[1,2]]
  iter2 = zigzagIterator(vecs)

  while len(iter1) > 0:
    print(iter1.next())

  while len(iter2) > 0:
    print(iter2.hasNext())
    print(iter2.next())
```

#421.**simply path**

描述        
给定一个文件的绝对路径(Unix-style)，请进行路径简化。

Unix中, . 表示当前目录, .. 表示父目录。

结果必须以 / 开头，并且两个目录名之间有且只有一个 /。最后一个目录名(如果存在)后不能出现 / 。你需要保证结果是正确表示路径的最短的字符串。

你是否考虑了 路径为 "/../" 的情况？在这种情况下，你需返回"/"。

此外，路径中也可能包含双斜杠'/'，如 "/home//foo/"; 在这种情况下，应该忽略多余的斜杠，返回 "/home/foo"。

输入: "/a/./../../c/"     
输出: "/c"              
解释: "/" 没有上级目录, "/../" 的结果就是 "/".

思路：        
split path by "/";      
if not "." or "", push it to stack;         
if "." or "", ignore;      
if "..", pop the top one from the stack if the stack is not empty;         
join the char in the stack by "/" -> result

```
def simplified_path(path):
  path = path.split("/")
  print(path)
  stack = []
  for char in path:
    if char == "..":         # 注意这里的各条件顺序，必须将该条写在最前面，如果调换if， elif这两个条件，会把".."也加到stack里面
      if len(stack):  # 如果stack不为空则pop当前目录，为空则pass
        stack.pop()
    elif char != "." and char != "":
      stack.append(char)


  return "/" + "/".join(stack)
    
path = "/a/./../../c/"
simplified_path(path)
```

#575.**decode string**

给出一个表达式 s，此表达式包括数字，字母以及方括号。在方括号前的数字表示方括号内容的重复次数(括号内的内容可以是字符串或另一个表达式)，请将这个表达式展开成一个字符串。

输入: S = abc3[a]    
输出: "abcaaa"

```
def decode_string(s):

  stack = []
  for ch in s:
    if ch != "]":
      stack.append(ch)
      continue       # 这里的continue很重要
    
    # cope with str inside brackets
    strs = []
    while stack and stack[-1] != "[":
      strs.append(stack.pop())
    
    # skip "["
    stack.pop()

    # cope with numbers, 注意这里的repeat和base的技巧
    repeats = 0
    base = 1
    while stack and stack[-1].isdigit():
      repeats += (ord(stack.pop()) - ord ("0")) * base
      base *= 10
    
    # push the repeat str into stack
    stack.append("".join(reversed(strs)) * repeats)

  return "".join(stack)


s = "ab10[cd]3[e]f"
decode_string(s)
```

#40. **用栈实现队列**
描述         
正如标题所述，你需要使用两个栈来实现队列的一些操作。

队列应支持push(element)，pop() 和 top()，其中pop是弹出队列中的第一个(最前面的)元素。

pop和top方法都应该返回第一个元素的值。

```
class MyQueue:

  def __init__(self):
    self.stack1 = []
    self.stack2 = []
  
  def push(self, e):
    self.stack1.append(e)
  
  def top(self):
    if len(self.stack1):
      return self.stack1[0]
    return None
  
  def pop(self):
    while self.stack1:
      element = self.stack1.pop()
      self.stack2.append(element)
    val = self.stack2.pop()
    while self.stack2:
      self.stack1.append(self.stack2.pop())
    return val

q = MyQueue()
for i in range(10):
  q.push(i)
print(q.stack1)
print(q.pop())
```

#546. **用接口实现队列**

```
class Node:
  def __init__(self,val):
    self.val = val
    self.next = None
    self.prev = None

class InterfaceQueue:
  def push(self, element):
    pass
  def pop(self):
    pass
  def top(self):
    pass

class MyQueue(InterfaceQueue):
  def __init__(self, first, last):
    self.first = first
    self.last = last
    self.first.next = self.last
    self.last.prev = self.first
  
  def push(self, val):
    new_node = Node(val)
    if self.last == None:
      self.first = new_node
      self.last = new_node
    
    else:
      self.last.next = new_node
      new_node.prev = self.last
      self.last = new_node
    return val
  
  def pop(self):
    if self.first == None:
      return None
    else:
      val = self.first.val
      self.first = self.first.next
      if self.first != None:
        self.first.prev = None
      else:
        self.last = None
      return val

  def top(self):
    """
    return the first element.
    """
    if self.first != None:
      return self.first.val
    return None
  
  def isEmpty(self):
    return self.first == None

if __name__ == "__main__":
  s = MyQueue(Node(0), Node(1))
  s.push(2)
  s.push(3)
  while not s.isEmpty():
    print(s.pop())
```

#71. **实现栈**

```
# 用双向链表实现
class Node:
  def __init__(self,val):
    self.val = val
    self.prev = None
    self.next = None

class myStack:
  def __init__(self,head, tail):
    self.head = head
    self.tail = tail
    self.head.next = self.tail
    self.tail.prev = self.head
  
  def isEmpty(self):
    return self.tail == None
  
  def push(self, val):
    new_node = Node(val)
    if self.tail == None:
      self.tail = new_node
      self.head = self.tail
    else:
      self.tail.next = new_node
      new_node.prev = self.tail
      self.tail = self.tail.next
      self.tail.next = None
    return val
  
  def pop(self):
    if self.tail == None:
      return None
    # 要考虑特殊情况：只有一个Node
    if self.tail.prev != None:       
      val = self.tail.val             
      self.tail = self.tail.prev
      self.tail.next = None
      return val
    else:
      val = self.tail.val
      self.tail = None
      self.head = None
      return val

  
  def top(self):
    if self.tail == None:
      return None
    return self.tail.val

if __name__ == "__main__":
  s = myStack(Node(0), Node(1))
  s.push(2)
  s.push(3)
  while not s.isEmpty():
    print(s.pop())
```

```
# implement stack by double queue
from collections import deque
class MyStack:

  def __init__(self):
    self.queue1 = deque()
    self.queue2 = deque()

  def push(self, x):
    self.queue1.append(x)
  
  def pop(self):
    for _ in range(len(self.queue1) - 1):   # 不包括最后一个element
      val = self.queue1.popleft()
      self.queue2.append(val)

    val = self.queue1.popleft()    # 这是最后一个element，应该pop出来的
    self.queue1, self.queue2 = self.queue2, self.queue1      # 交换
    return val

  def top(self):       # 最上面的值
    val = self.pop()   # 用pop取出这个值
    self.push(val)     # 再用push把这个值还回原处
    return val
  
  def isEmpty(self):
    return not self.queue1     
```

#423. **有效的括号序列**

描述         
给定一个字符串所表示的括号序列，包含以下字符： '(', ')', '{', '}', '[' and ']'， 判定是否是有效的括号序列。

括号必须依照 "()" 顺序表示， "()[]{}" "[()]{}"是有效的括号，但 "([)]" 则是无效的括号。      
O(n)的时间，n 为括号的个数

```
# implement by deque
def valid_parenthesis(A):

  a = deque()
  b = deque()
  for p in A:
    a.append(p)
  
  d = {"(":")", "[":"]","{":"}"}
  while len(a) > 0:
    val = a.popleft()
    if val in d:
      b.append(val)
    else:
      if b:
        if d[b[-1]] == val:
          b.pop()
        else:
          return False
      else:
        return False
  if len(b) == 0:
    return True
  else:
    return False

if __name__ == "__main__":
  A1 = "{()[]}"
  A2 = "())"
  A3 = "][()"

  print(valid_parenthesis(A1))
  print(valid_parenthesis(A2))
  print(valid_parenthesis(A3))
```
