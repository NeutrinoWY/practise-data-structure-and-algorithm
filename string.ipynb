# 1. **【反转整数】**
反转一个只有3位数的整数

```
def reverse_num(number):
  assert isinstance(number,int)
  assert number < 1000

  digits = number % 10
  tens = number % 100 // 10
  hundreds = number // 100

  reverse_num = digits*100 + tens*10 + hundreds

  return reverse_num

if __name__ == "__main__":
  #reverse_num('123')
  print(reverse_num(123))
  print(reverse_num(120))
  #reverse_num(1203)
```

```
def leapYear(year):
  assert isinstance(year,int)

  if year%4==0 and year%100!=0:
    return True
  elif year%400 == 0:
    return True
  else:
    return False

if __name__ == "__main__":
  print(leapYear(2017))
```

3. 给你一个整数n. 从 1 到 n 按照下面的规则打印每个数：

如果这个数被3整除，打印fizz.
如果这个数被5整除，打印buzz.
如果这个数能同时被3和5整除，打印fizz buzz.
如果这个数既不能被 3 整除也不能被 5 整除，打印数字本身。

样例
比如 n = 15, 返回一个字符串数组：

[
  "1", "2", "fizz",
  "4", "buzz", "fizz",
  "7", "8", "fizz",
  "buzz", "11", "fizz",
  "13", "14", "fizz buzz"
]

```
def printFizzBuzz(n):
  assert isinstance(n, int)

  result = [0]*n
  for i in range(1, n+1):
    if i % 15 == 0:
      result[i-1] = "fizz buzz"
    elif i % 3 == 0:
      result[i-1] = "fizz"
    elif i % 5 == 0:
      result[i-1] = "buzz"
    else:
      result[i-1] = str(i)
  
  return result

printFizzBuzz(15)
```

# **4. 【jump game】- 【贪心算法】【动态规划】 - 模板题**

给出一个非负整数数组，你最初定位在数组的第一个位置;
数组中的每个元素代表你在那个位置可以跳跃的最大长度;
判断你是否能到达数组的最后一个位置。

样例 1

输入 : [2,3,1,1,4]
输出 : true

样例 2

输入 : [3,2,1,0,4]
输出 : false

```
def canJump(A):
  assert A != []

  p = 0
  ans = 0
  for item in A[:-1]:
    ans = max(ans, p + item)
    if ans <= p:    #key point: 贪心法
      return False
    p += 1
  return True

if __name__ == "__main__":
  print(canJump([3,2,1,0,4]))
  print(canJump([2,3,1,1,4]))
```

给你一个数组和两个索引，交换下标为这两个索引的数字

```
def swap_elements(l,idx1,idx2):

  assert len(l) >= 2 and isinstance(idx1,int) and isinstance(idx2, int)

  temp = l[idx1]
  l[idx1] = l[idx2]
  l[idx2] = temp

  return l

swap_elements([1,2,3,4],0,2)
```

```
def fibonacci(n):
  a = 0
  b = 1
  for i in range(n - 1):
    a,b = b, a+b
  return a

fibonacci(3)
```

如果方程只有一个根，就返回一个包含一个跟的数组/列表。
如果方程没有根，就返回一个空数组/列表。

输入：a = 1, b = -3, c = 2
输出：[1,2]
解释：方程有两个根，返回 [1,2] 且第一个数应比第二个数小。

```
import math
def solve_quadratic_equation(a,b,c):
  assert a != 0
  if b**2 - 4*a*c < 0:
    return []
  else:
    return sorted(list(set([(math.sqrt(b**2 - 4*a*c)-b)/2*a, (-1*math.sqrt(b**2 - 4*a*c)-b)/2*a ])))
  
solve_quadratic_equation(1,-3,2)
```

```
# 面试题不能直接这样调用API
def strToInt(s):
  if s[0] == "-":
    return -1*int(s[1:])
  else:
    return int(s)
  
strToInt("-120")
```

解题思路：
1. 先考虑正整数
2. 再考虑负整数
3. 考虑、测试边界条件（字符串为空，字符串为‘0’）

```
def strToInt(s):

  if len(s)==0:
    return 0

  if s[0] == "-":
    start = 1
  else:
    start = 0
  result = 0
  for i in range(start, len(s)):
    num = ord(s[i]) - ord('0')       # point 1
    if s[0] != "-":
      result = result * 10 + num     # point 2
    ## s = "123"
    ## i=0, num=1, result=1
    ## i=1, num=2, result= 1*10 + 2 = 12
    ## i=2, num=3, result = 12*10+3 = 123
    else:
      result = 10*result - num        # point 3
      print(result)
    ## s = "-123"
    ## i=1, num=1, result=-1
    ## i=2, num=2, result=-10-2 = -12
    ## i=3, num=3, result = -120-3 = -123
  
  return result


strToInt("-1230")
```

# 958. **【数据流回文字符串】**
一个数据流进来，每次一个小写字母，当前数据流的排列是否能组成回文串

```
def streamPalindrome(s):
  n = len(s)
  assert n != 0
  result = []
  letter_dict = {}
  #substring = ""
  for i in range(n):
    char = s[i]
    #substring += char
    if char in letter_dict:
      letter_dict[char] += 1
    else:
      letter_dict[char] = 1
    modes = [v % 2 for v in letter_dict.values()]
    if sum(modes) > 1:        # the number of odds
      result.append(0)
    else:
      result.append(1)
  return result


streamPalindrome('abcbc')
```

# 1819. **【最长子字符串】**
描述
给定一个长度为NN且只包含a和b的字符串SS。你需要找出最长的子串长度，使得其中不包含三个连续的字母。即，找出不包含aaa或bbb的最长子串长度。注意SS是其本身的子串。

输入: "baaabbabbb"
输出: 7
说明: "aabbabb"是最长符合条件的子串

Follow up questions: 1. 包括26个字母； 2. 不包含k个连续字母的子串长度

```
def longestSemiAlternatingSubstring(s):

  if s is None:return None
  if len(s) < 3:return len(s)

  start = 0
  end = 1
  max_len = 0
  note = 1
  while end < len(s):
    if s[end] == s[end-1]:
      note += 1
      if note == 3: 
        start = end - 1
        note = 2
    else:
      note = 1
    max_len = max(max_len, (end - start + 1))
    end += 1
  return max_len

longestSemiAlternatingSubstring("aaabaaa")
```

# **146.【大小写转换】**
将一个字符串中的小写字母转换为大写字母。不是字母的字符不需要做改变。

输入: str = "aBc"； 输出: "ABC"

输入: str = "abC12"； 输出: "ABC12"



时间复杂度O(n), 空间复杂度O(n)

```
def lowerToUpper(s):
  n = len(s)
  new_string = ""
  for i in range(n):
    char = s[i]
    if char >= "a" and char <= "z":          #用此条件去检测是不是小写字母
      char = chr(ord(s[i]) - ord("a") + ord("A"))     #小写字母可以用这种方法转换成大写字母，但是大写字母和数字不能用
    new_string += char
  return new_string

lowerToUpper("abC12")
```

# 1540. **字符串包含**
给两个字符串 S 和 T, 判断 S 能不能通过删除一些字母(包括0个)变成 T.

输入: S = "lintcode" 和 T = "lint";  
输出: true

注意：问题不是S中是否有连续的子字符串和T是一样的，因为可以删除不连续的一些字母使得剩下来的字母连起来和t一样

```
# 子字符串与t一样【不符合题意】
def canConvert(s,t):
  if s == None: 
    if t == None:
      return True
    return False

  for start in range(0, len(s) - len(t)+1):
    substring = s[start:start+len(t)]
    if substring == t:
      return True
  return False

canConvert("helloe","elle")
```

思路：对t中的每一个character，去s中找相同的，如遇到一个，count加1， 再看t中的下一个character，从s中上个相同的character后面接着找

时间复杂度O(n), n=len(s) + len(t), 空间复杂度O(1)

```
def canConvert(s, t):
  if s == None: 
    if t == None:
      return True
    return False

  count = 0
  start = 0
  for item in t:
    for i in range(start, len(s)):
      if s[i] == item:
        start = i + 1
        count += 1
        break
  if count == len(t):
    return True
  else:
    return False

canConvert('a','aa')
```

# 936. **【首字母大写】** 
输入一个英文句子，将每个单词的第一个字母改成大写字母

输入: s =  "i jidls    mdijf  i  lsidj  i p l    "；  
输出: "I Jidls    Mdijf  I  Lsidj  I P L    "

注意：单词之前的space有几个就需要保留几个！

```
def capitalizeFirstLetter(s):
  if s == None:return None
  if s == "":return ""

  l = s.split()  # split会去掉space
  new_list = []
  for sub in l:
    new = chr(ord(sub[0]) - ord('a') + ord('A')) + sub[1:]
    new_list.append(new)
  return ' '.join(new_list)
  
capitalizeFirstLetter("i jidls mdijf i lsidj i p l ")
```

注意：单词之前的space有几个就需要保留几个！
时间复杂度O(N), 空间复杂度O(N)

```
def capitalizeFirstLetter(s):
  #return s.title()
  if s == None:return None
  if s == "":return ""

  new = ""
  cap = 1   # 用一个变量cap来标记什么时候该大写
  for i in range(len(s)):
    char = s[i]
    if char != " " and cap == 1:        # char不为space,同时cap=1,则大写
      char = chr(ord(char) - ord('a') + ord('A'))
      cap = 0
    if char == " ":      
      cap = 1
    new += char
  return new


capitalizeFirstLetter("i jidls mdijf i lsidj i p l ")
```

给定一个字符串，逐个翻转字符串中的每个单词

时间复杂度O(n), 空间复杂度O(n)

```
def reverseStr(s):
  if s == None: return None
  if s == "": return ""

  words = s.split()
  return ' '.join(words[-1::-1])

reverseStr("the sky is blue,  ")
```
