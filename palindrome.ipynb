**错误的生成矩阵的方法**

```python
# 这样生成矩阵，外层是将[0, 0, 0]复制，但位置还是原来这个内层list的位置
m = [[0] * 3] * 3

# 所以会改变每一排第一个元素的值
m[0][0] = 1
print(m)
```

**正确生成矩阵的方法**

```python
# 这样能控制每一个位置的元素
m = [[0] * 3 for _ in range(3)]

# 这样就能只改变第一行第一列的元素
m[0][0] = 1
print(m)
```

**13. 字符串查找**

对于一个给定的 source 字符串和一个 target 字符串，你应该在 source 字符串中找出 target 字符串出现的第一个位置(从0开始)。如果不存在，则返回 -1。

思路： 这里涉及到source和target的index下标，要非常注意是否会出现index out of range的问题，还需要注意是否将每个字符都搜索到。          
情况1： source 与 target 长度相等；    
情况2： source 长度大于 target；     
情况3:  source 长度小于 target。

```python
def search_string(source, target):
  if len(source) < len(target):
      return -1

  elif len(source) == len(target):
    if source == target:
      return 0
    else:
      return -1

  else:
    length = len(target)
    for i in range(len(source) - length + 1):      # range里需要加1将source中的最后一个字符遍历到，因为可能target就只有1个字符
      if source[i : i + length] == target:
        return i
    return -1

if __name__ == "__main__":
  print(search_string("hello", "ll"))
  print(search_string("he", ""))
  print(search_string("", "ll"))
  print(search_string("abcde", "e"))

  
```

**627. 最长回文串**

给出一个包含大小写字母的字符串。求出由这些字母可以构建的最长的回文串的长度是多少。数据是大小写敏感的，也就是说，"Aa" 并不会被认为是一个回文串。

输入 : s = "abccccdd"     
输出 : 7     
说明 : 一种可以构建出来的最长回文串方案是 "dccaccd"

思路： 回文串的字符特点，对于偶数个字符，每个字符都有偶数个；对于奇数回文串，其中一个字符可以是奇数，其他字符都必须是偶数。      
可以用一个dictionary记录字符及其个数；对每一个字符，取最大的偶数，为了使回文串最长，最后还可以加1使其成为奇数回文串。

特殊情况：    
1. 字符串为空； ->  0       
2. 字符串只有一个字符；“a" -> 1
3. 字符串有偶数个同一字符，如“aaaa” -> 4
4. 字符串有奇数个同一字符，如"aaaaa" -> 5

```python
def longest_palindrome(string):
  if string is None or len(string) == 0:
    return 0
  
  mp = {}
  for char in string:
    if char not in mp:
      mp[char] = 1
    else:
      mp[char] += 1

  letters = mp.keys()
  result = 0
  odd = 0     # 用odd来记录string中的字符个数是否都是偶数个，如果都是偶数个则为0
  for l in letters:
    if mp[l] % 2 == 0:
      result += mp[l]
    else:
      result += mp[l] - 1
      odd += 1
  
  if odd == 0:    #如果所有字符都是偶数个，则不能在result上加1，因为已经用上了string中的所有字符了
    return result
  else:           #如果有奇数的字符，则可以加1
    return result + 1

if __name__ == "__main__":

  s = "NTrQdQGgwtxqRTSBOitAXUkwGLgUHtQOmYMwZlUxqZysKpZxRoehgirdMUgy"
  print(longest_palindrome(s))
  s1 = "aaaa"
  print(longest_palindrome(s1))
  s2 = "ddddd"
  print(longest_palindrome(s2))
  s3 = "abababab"
  print(longest_palindrome(s3))
```
